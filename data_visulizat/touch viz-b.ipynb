```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
### The avarage people survived based on thier class 
```

```python

sns.get_dataset_names()
tit = sns.load_dataset("titanic")
tit.head(20)
```

```python
survive_class = tit.groupby("class", as_index=False).mean()
sns.barplot(x="class",y="survived",data=survive_class)
```

```python
### The avarage people survived based on thier a lone or not  
```

```python
survive_class = tit.groupby("sex", as_index=False).mean()
sns.barplot(x="sex",y="survived",data=tit)
```

```python
### The avarage people survived based on thier gender
```

```python
survive_class = tit.groupby("alone", as_index=False).mean()
sns.barplot(x="alone",y="survived",data=tit)
```

```python
sns.relplot(x="who",y="class", hue="survived", data=tit)
```
